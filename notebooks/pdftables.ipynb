### This notebook demonstrates how to use pdfplumber to extract data from tables in PDF to a pandas dataframe.

```python
import pdfplumber
```

```python
import dataiku
from dataiku import pandasutils as pdu
import pandas as pd
```

```python
## Getting the path of PDF file saved in Dataiku folder
folder = dataiku.Folder("waste")
folder_info = folder.get_info()
folder.list_paths_in_partition()
```

```python
paths = folder.list_paths_in_partition()
paths[0]
```

```python
report = folder_info['path']+paths[0]
report
```

```python
### Extracting first table
with pdfplumber.open(report) as pdf:
    page = pdf.pages[0]    
    table = page.extract_table()
    df = pd.DataFrame(table, columns=table[0])
    new_header = df.iloc[1] #grab the first row for the header
    df = df[2:] #take the data less the header row
    df.columns = new_header #set the header row as the df header
    df['Year'] = 2022
```

```python
### Extracting second table
with pdfplumber.open(report) as pdf:
    page = pdf.pages[0]    
    tables = page.extract_tables()
    df1 = pd.DataFrame(tables[1], columns=tables[1][0])
    new_header = df1.iloc[1] #grab the first row for the header
    df1 = df1[2:] #take the data less the header row
    df1.columns = new_header #set the header row as the df header
    df1['Year'] = 2021
```

```python
df1
```

```python
### Appending the two tables together
df = df.append(df1, ignore_index=True)
```

```python
df
```

```python
### I only needed the first two tables on the first page hence I didn't extract further
### There is one table that is splitted across two pages and will need some data wrangling if we want that data
### Looking at the tables on second page
with pdfplumber.open(report) as pdf:
    page = pdf.pages[1]    
    tables = page.extract_tables()
    
```

```python
with pdfplumber.open(report) as pdf:
    page = pdf.pages[1]    
    table = page.extract_table()
    df = pd.DataFrame(table, columns=table[0])
```
