```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

url = 'https://www.chope.co/singapore-restaurants/pages/100-best-restaurants-singapore?lang=en_US'

```

```python
# Send a GET request to the URL
response = requests.get(url)

# Parse HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all restaurant names and locations
restaurants = soup.find_all('a', class_='restaurantLink')
locations = soup.find_all('div', class_='list__label--location')
cuisines = soup.find_all('div', class_='list__label--cuisine')
descriptions = soup.find_all('div', class_='list__label--description')
```

```python
food_list = []
```

```python
# Extract and print restaurant names and locations
for i in range(0,100):
    name = restaurants[2*i-1].get_text()
    url_link = restaurants[2*i-1]['href']
    location = locations[i].get_text()
    cuisine = cuisines[i].get_text()
    description = descriptions[i].get_text()
    #print(i)
    food_list.append((name, url_link, location, cuisine, description))
```

```python
df = pd.DataFrame(np.array(food_list))
```
